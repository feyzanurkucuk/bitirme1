```python
import cv2
import os

def extract_frames(video_path, output_folder):
    # Open the video file
    cap = cv2.VideoCapture('video/video_no6.mp4')

    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get the frames per second (fps) and frame count of the video
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    print(f"Video FPS: {fps}")
    print(f"Total Frames: {frame_count}")

    # Loop through each frame and save it as an image
    for frame_number in range(frame_count):
        ret, frame = cap.read()

        if not ret:
            print("Error reading frame.")
            break

        # Save the frame as an image
        frame_filename = os.path.join(output_folder, f"frame_no5_{frame_number + 1:04d}.jpg")
        cv2.imwrite(frame_filename, frame)

    # Release the video capture object
    cap.release()

if __name__ == "__main__":
    # Replace 'input_video.mp4' with the path to your video file
    input_video_path = 'input_video.mp4'

    # Replace 'output_frames' with the folder where you want to save the frames
    output_frames_folder = 'output_frames'

    extract_frames(input_video_path, output_frames_folder)
```

```python
import os

def rename_images(folder_path, new_prefix='image'):
    # Get a list of all files in the folder
    files = os.listdir(folder_path)

    # Filter only image files (you may want to adjust the list of valid extensions)
    image_files = [file for file in files if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

    # Iterate over image files and rename them
    for i, old_name in enumerate(image_files):
        # Get file extension
        _, file_extension = os.path.splitext(old_name)

        # Generate new file name with the specified prefix and index
        new_name = f'frame_no_{i + 1}{file_extension}'

        # Construct full paths
        old_path = os.path.join(folder_path, old_name)
        new_path = os.path.join(folder_path, new_name)

        # Rename the file
        os.rename(old_path, new_path)

        print(f'Renamed: {old_name} -> {new_name}')

# Example usage
folder_path = 'output_frames_no'
rename_images(folder_path, new_prefix='')
```

```python
import cv2
import os

def uniform_subsampling_and_save(video_path, output_folder, interval):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    video_capture = cv2.VideoCapture(video_path)
    frame_count = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))

    for i in range(0, frame_count, interval):
        video_capture.set(cv2.CAP_PROP_POS_FRAMES, i)
        ret, frame = video_capture.read()
        if ret:
            # Construct the output filename
            output_filename = os.path.join(output_folder, f'frame9_{i//interval}.jpg')

            # Save the frame as an image file
            cv2.imwrite(output_filename, frame)

    video_capture.release()

if __name__ == "__main__":
    video_path = 'video/video_no3.mp4'
    output_folder = 'output_frames_no_3'
    interval = 4  # Adjust the interval based on your preferences

    uniform_subsampling_and_save(video_path, output_folder, interval)
```

```python
from PIL import Image
import os

def crop_images(input_folder, output_folder, target_resolution=(720, 720)):
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Loop through each file in the input folder
    for filename in os.listdir(input_folder):
        input_path = os.path.join(input_folder, filename)

        # Check if the file is an image (you may need to adjust the list of valid image extensions)
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            # Open the image
            with Image.open(input_path) as img:
                # Calculate cropping dimensions
                original_width, original_height = img.size
                crop_top = int((original_height - target_resolution[1]) / 2)
                crop_bottom = original_height - crop_top

                # Crop the image
                cropped_img = img.crop((0, crop_top, original_width, crop_bottom))

                # Resize the cropped image to the target resolution
                resized_img = cropped_img.resize(target_resolution)

                # Save the resized and cropped image to the output folder
                output_path = os.path.join(output_folder, filename)
                resized_img.save(output_path)

if __name__ == "__main__":
    input_folder = 'output_frames_no_3'
    output_folder = 'output_frames_no1'
    target_resolution = (480, 480)

    crop_images(input_folder, output_folder, target_resolution)
```

```python
from PIL import Image
import os

def resize_and_crop_images(input_folder, output_folder, target_width, target_height):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        # Open the image
        img = Image.open(input_path)

        # Resize the image while maintaining its aspect ratio
        img.thumbnail((target_width, target_height))

        # Calculate the crop box to center the image
        left = (img.width - target_width) // 2
        top = (img.height - target_height) // 2
        right = (img.width + target_width) // 2
        bottom = (img.height + target_height) // 2

        # Crop the image
        img = img.crop((left, top, right, bottom))

        # Save the resized and cropped image
        img.save(output_path)

if __name__ == "__main__":
    input_folder = 'output_frames_no_3'
    output_folder = 'output_frames_no1'
    target_width = 480
    target_height = 480

    resize_and_crop_images(input_folder, output_folder, target_width, target_height)
```

```python
import numpy as np
import cv2
def preprocess_image(image_path):
    # Load and preprocess the image using OpenCV
    image = Image.open(image_path).convert('L')  # convert to grayscale

    # Convert the PIL image to a numpy array
    image_np = np.array(image)
    if len(image_np.shape) == 3 and image_np.shape[2] == 3:
        image_np = cv2.cvtColor(image_np, cv2.COLOR_BGR2GRAY)

    # Normalize and resize the image
    normalized_image = image_np / 255.0
    resized_image = cv2.resize(normalized_image, (128, 64))
    resized_image = np.expand_dims(resized_image, axis=0)
    return resized_image
```

```python
import os
from PIL import Image
import json

def load_images_and_annotations(image_folder, label_folder):
    images = []
    labels = []
    points = []

    # Iterate through each image file in the image folder
    for filename in os.listdir(image_folder):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            image_path = os.path.join(image_folder, filename)
            image = Image.open(image_path)
            images.append(image)

            # Load corresponding label from the label folder
            label_filename = os.path.splitext(filename)[0] + '.json'
            label_path = os.path.join(label_folder, label_filename)

            if os.path.exists(label_path):
                with open(label_path, 'r') as label_file:
                    label_data = json.load(label_file)
                    # Assuming your label structure is similar to the provided JSON
                    label = label_data['shapes'][0]['label']
                    point = label_data['shapes'][0]['points']
                    labels.append(label)
                    points.append(point)
            else:
                # Handle the case where there's no corresponding label file
                labels.append(None)

    return images, labels,points

image_folder = 'dataset/images'
label_folder = 'dataset/labels'

images, labels,points = load_images_and_annotations(image_folder, label_folder)

```

```python
def convert_to_yolo_format(xmin, ymin, xmax, ymax, image_width, image_height, class_label):
    # Calculate normalized coordinates
    x_center = (xmin + xmax) / (2.0 * image_width)
    y_center = (ymin + ymax) / (2.0 * image_height)
    width = (xmax - xmin) / image_width
    height = (ymax - ymin) / image_height

    # YOLO format: class_label x_center y_center width height
    yolo_format = f"{class_label} {x_center} {y_center} {width} {height}"

    return yolo_format
```
